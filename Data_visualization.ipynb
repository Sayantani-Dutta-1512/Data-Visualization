```python
import pandas as pd
import numpy as np
ipl = pd.read_csv('ipl.csv')
ipl.head(10)
```

```python
ipl.isnull().sum()
```

```python
x = ipl["umpire3"].median()
ipl["umpire3"].fillna(x, inplace = True)
print(x)
print(ipl)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(ipl.isnull())
```

```python
sns.barplot(x="city", y="result", data=ipl)
```

```python
sns.jointplot(x="city", y="result", data=ipl)
```

```python
sns.pairplot(ipl, hue="city")
```

```python
import matplotlib.pyplot as plt
from scipy import stats
x = 'season'
y = 'result'
plt.scatter(ipl[x], ipl[y])
plt.show()
```

```python
result = ipl["result"]
plt.plot(result, color = 'red', label='result')
plt.xlabel('season')
plt.ylabel('result')
plt.title('result vs season')
plt.legend()
plt.show()
```

```python
sns.displot(x ='result',data=ipl)
```

#World cup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
world_cup= pd.read_csv('world_cup.csv')
world_cup.head(10)
```

```python
world_cup.isnull().sum()
```

```python
x = world_cup["Attendance"].median()
world_cup["Attendance"].fillna(x, inplace = True)
print(x)
print(world_cup)
```

```python
sns.heatmap(world_cup.isnull())
```

```python
sns.barplot(x="Year", y="City", data=world_cup)
```

```python
sns.jointplot(x="City", y="Year", data=world_cup)
```

```python
sns.pairplot(world_cup, hue="City")
```

```python
import matplotlib.pyplot as plt
from scipy import stats
x = 'Attendance'
y = 'Year'
plt.scatter(world_cup[x], world_cup[y])
plt.show()
```

```python
Attendance = world_cup["Attendance"]
plt.plot(Attendance, color = 'red', label='result')
plt.xlabel('City')
plt.ylabel('Attendance')
plt.title('Attendance vs City')
plt.legend()
plt.show()
```
